```python
# data from Weather Underground
from datetime import datetime
import requests, json, csv

start_date = "2014-01-01"
stop_date = "2014-12-31"

start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")

data = []
from datetime import timedelta
while start <= stop:
    date_formatted = str(start.year) + '{0:02d}'.format(start.month) + '{0:02d}'.format(start.day)
    url = 'http://api.wunderground.com/api/f5e2220e584de024/history_%s/q/PA/Pittsburgh.json' % date_formatted    
    print url
    r = requests.get(url)
    response = r.json()
    daily_summary = response['history']['dailysummary'][0]
    data.append(daily_summary)
    
    start = start + timedelta(days=1)  # increase day one by one

for row in data:
    for k, v in row['date'].iteritems():
        row[k] = v

with open('2014-weather.csv', 'wb') as f:
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    writer.writeheader()
    for row in data:
        writer.writerow(row)
    print 'Wrote', str(len(data)), 'rows'
```
